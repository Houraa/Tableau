```python
import pandas as pd
import numpy as np
import seaborn as sb

data = pd.read_csv('/content/drive/My Drive/Vis Project/players_22_v2.csv')
```

```python
data
```

```python
features = {}
inx1 = data.columns.get_loc('attacking_finishing')
inx2 = data.columns.get_loc('goalkeeping_speed')

features['nationality_name'] = data['nationality_name']
for i in range(inx1, inx2):
    features[data.columns[i]] = data[data.columns[i]]

features = pd.DataFrame(features)
features = features.groupby('nationality_name').mean()
features
```

```python
features['attacking_finishing'].keys()
```

```python
import operator

nationality_tops = {}
nationality_tops['nationality_name'] = features['attacking_finishing'].keys()

# print(len(nationality_tops['nationality_name']))
top1 = []
top2 = []
top3 = []
top1_names = []
top2_names = []
top3_names = []
for country in nationality_tops['nationality_name']:
  country_tmp = {}

  for feature in features.columns:
    country_tmp[feature] = features[feature][country]

  sorted_country_tmp = dict(sorted(country_tmp.items(), key=operator.itemgetter(1),reverse=True))
  country_list = list(sorted_country_tmp.items())
  top1.append(country_list[0][1])
  top2.append(country_list[1][1])
  top3.append(country_list[2][1])
  top1_names.append(country_list[0][0])
  top2_names.append(country_list[1][0])
  top3_names.append(country_list[2][0])

nationality_tops['Top 1'] = top1
nationality_tops['Top 2'] = top2
nationality_tops['Top 3'] = top3
nationality_tops['Top 1 Name'] = top1_names
nationality_tops['Top 2 Name'] = top2_names
nationality_tops['Top 3 Name'] = top3_names

nationality_tops = pd.DataFrame(nationality_tops)
```

```python
# data.join(nationality_tops)
new_data = pd.merge(data, nationality_tops, on='nationality_name', how='outer')
new_data = new_data.sort_values(by=['index'])
new_data
```

```python
new_data['nationality'] = new_data['nationality_name']
```
